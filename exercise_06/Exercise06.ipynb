```python
from exercise06 import BookData
import pandas as pd

url_list = ["https://www.gutenberg.org/files/2701/2701-0.txt", "https://www.gutenberg.org/files/209/209-0.txt",
            "https://www.gutenberg.org/files/46/46-0.txt", "https://www.gutenberg.org/files/42108/42108-0.txt", "https://www.gutenberg.org/files/74/74-0.txt", "https://www.gutenberg.org/cache/epub/514/pg514.txt", "https://www.gutenberg.org/cache/epub/11030/pg11030.txt", "https://www.gutenberg.org/files/4300/4300-0.txt"]

filenames = ["Moby Dick", "The Turn of the Screw",
             "A Christmas Carol", "The Slang Dictionary", "THE ADVENTURES OF TOM SAWYER", "LITTLE WOMEN", "Incidents in the Life of a Slave Girl", "Ulysses"]


book_data = BookData(url_list, filenames)

book_data.multi_download()

hardest_books = book_data.hardest_read()

d = {'Book Title' : hardest_books.keys(), 'Vowels' : hardest_books.values()}

df = pd.DataFrame(d).sort_values('Vowels', ascending=False)

display(df)
df.plot(kind="bar", x='Book Title', y='Vowels', figsize=(11,5))
```
